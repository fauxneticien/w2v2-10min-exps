```python
import glob
import pandas as pd
```

```python
results_df = pd.concat([
    pd.read_csv(p).assign(path=p) for p in glob.glob("tmp/results/*/*/results.csv", recursive=True) 
]) \
.assign(
    lang = lambda x: x.path.str.extract("(ENGcc|ENG|FRY|GOS)"),
    lm   = lambda x: x.path.str.extract("(4gram_FULL|\dgram_[\d|_]+)")
) \
.fillna({'lm':'None'})

results_df
```

```python
import numpy as np
import math

# results_df[['lang', 'lm', 'wer_dev', 'wer_test', 'cer_dev', 'cer_test']] \

results_df[['lang', 'lm', 'wer_dev', 'wer_test', 'cer_dev', 'cer_test']] \
    .melt(id_vars=['lang', 'lm']) \
    .assign(
        measure = lambda x: [ s.split('_')[0] for s in x.variable.to_list() ],
        dataset = lambda x: [ s.split('_')[1] for s in x.variable.to_list() ]
    ) \
    .groupby(['lang', 'lm', 'dataset', 'measure']) \
    .agg(
        Mean=('value', np.mean),
        SD=('value', np.std)
    ) \
    .fillna(-1) \
    .reset_index() \
    .assign(
        stat = lambda x: [ f"{m * 100:.2f} ({(s * 100):.2f})".replace("(-100.00)", "(n.a.)") for (m, s) in zip(x.Mean, x.SD) ]
    ) \
    [['lang', 'lm', 'dataset', 'measure', 'stat']] \
    .pivot(index=['lang', 'lm'], columns=['dataset', 'measure'], values='stat')
```
